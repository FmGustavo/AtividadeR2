```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
df = pd.read_excel("Populacao_do_Brasil.xlsx")
```

```python
df
```

```python
df.rename(columns={"Ano": "X", "Populacao": "Y"}, inplace=True)
```

```python
X = df["X"].values.reshape(-1, 1)
Y = df["Y"].values
```

```python
print(df.columns)
```

```python
# Criar modelo de regressão linear
modelo = LinearRegression()
modelo.fit(X, Y)

# Coeficientes da equação da reta: Y = aX + b
a = modelo.coef_[0]
b = modelo.intercept_

# Previsão e cálculo do R²
Y_pred = modelo.predict(X)
r2 = r2_score(Y, Y_pred)
```

```python
# Exibir a equação e R²
print(f"Equação da reta: População = {a:.2f} * Ano + {b:.2f}")
print(f"Coeficiente de determinação (R²): {r2:.4f}")
```

```python
# Plotar gráfico
plt.scatter(X, Y, color='blue', label="Dados reais")
plt.plot(X, Y_pred, color='red', label="Reta de regressão")
plt.xlabel("Ano")
plt.ylabel("População")
plt.legend()
plt.show()
```

```python
anos = df["X"].to_numpy()        # Array dos anos
populacao = df["Y"].to_numpy()  # Array da população

print("Anos:", anos)
print("População:", populacao)
```
